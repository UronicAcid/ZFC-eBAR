```python
import argparse
import zfc
```

```python

normalization = 'median'
use_small_count = False
min_ctrl_counts = 10


inputdata = zfc.read_raw_count(
    '../A549_D0_vs_D35.count.txt',
    gene_colname='gene',
    guide_colname='guide',
    barcode_colname='barcode',
    ctrl_colname='ctrl',
    exp_colname='exp'
)


barresult, sgresult, gresult, train_data, lm = zfc.zfoldchange(
    inputdata,
    top_n_sgrna=None,
    top_n_gene=None,
    iteration=100,
    normalization='median'
)

zfc.write_bar_result(
    barresult,
    '_'.join([args['outprefix'], 'barcode.txt'])
)

zfc.write_sg_result(
    sgresult,
    '_'.join([args['outprefix'], 'sgrna.txt'])
)

zfc.write_g_result(
    gresult,
    '_'.join([args['outprefix'], 'gene.txt'])
)


zfc.zfc_plot(
    args['outprefix'], inputdata,
    barresult, sgresult, gresult,
    train_data, lm
)
```
